```python
rows=6
col=2    
for i in df:
    plt.figure(figsize=(22,50))
    fig=1
    plt.subplot(rows,col,fig)
    sns.distplot(df[i])
    fig=fig+1
    plt.subplot(rows,col,fig)
    sns.boxplot(df[i])
    fig=fig+1
    plt.show()
    
    
```

We now handle the outliers using IQR Based Filtering.

```python
row=6
col=2
fig=1
for i in df:
    print(i)
    
    #finding IQR
    percentile25 = df[i].quantile(0.25)
    percentile75 = df[i].quantile(0.75)

    #Finding the upper and lower limits
    Q3 = np.quantile(df[i], 0.75)
    Q1 = np.quantile(df[i], 0.25)
    iqr= Q3 - Q1
    upper_limit = percentile75 + 1.5 * iqr
    lower_limit = percentile25 - 1.5 * iqr 
    
    new_df = df[[df[i] < upper_limit] and [df[i] > lower_limit]]
    new_df
    
    plt.figure(figsize=(12,5))
    plt.subplot(rows,col,fig)
    sns.boxplot(df[i])
    fig=fig+1
    plt.show()
    
    plt.figure(figsize=(12,5))
    plt.subplot(rows,col,fig)
    sns.boxplot(new_df[i])
    fig=fig+1
    plt.show()
    
    
```

```python
#trimming outliers
new_df = df[df['Pregnancies'] < upper_limit]
new_df.shape

plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['Pregnancies'])
plt.subplot(2,2,2)
sns.boxplot(df['Pregnancies'])
plt.subplot(2,2,3)
sns.distplot(new_df['Pregnancies'])
plt.subplot(2,2,4)
sns.boxplot(new_df['Pregnancies'])
plt.show()
```
